```python
dir_path = "/content/drive/MyDrive/S&P Crypto Project/Transfer Entropy/TE result-2.xlsx"
result = pd.read_excel(dir_path,sheet_name = 1)
result['err1'] = (result['max1'] - result['min1'])/2
result['err2'] = (result['max2'] - result['min2'])/2
result
```

##**1.Gt_bitcoin**

```python
result2 = pd.read_excel(dir_path,sheet_name = 1)
result2['err1'] = (result2['max1'] - result2['min1'])/2
result2['err2'] = (result2['max2'] - result2['min2'])/2
result2
```

```python
%matplotlib inline
# a = [1, 3, 5, 7]
# b = [11, -2, 4, 19]

# plt.rcParams["figure.figsize"] = [12, 8]
# plt.rcParams["figure.autolayout"] = True

plt.figure(figsize=(12,8))

# plt.text(0.05, 0.02, 'X: Gt_ethereum Return'+'\n'+'Y: Bitcoin Return', fontsize = 18, verticalalignment='top')
plt.scatter(result2['lag'], result2['ETE1'], label = 'X -> Y',s=100)
plt.scatter(result2['lag'], result2['ETE2'], label = 'Y -> X',s=100)

# c = [1, 3, 2, 1]
plt.errorbar(result2['lag'], result2['ETE1'], yerr=result2['err1'], fmt="-o")
plt.errorbar(result2['lag'], result2['ETE2'], yerr=result2['err2'], fmt="-o")
plt.xlabel('lag (days)', fontsize=20)
plt.ylabel('Effective Transfer Entropy (ETE)', fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=15)

plt.legend(loc = 3,bbox_to_anchor=(0.05, 0.05), fontsize=18)
plt.annotate('X: Gt_bitcoin Return'+'\n'+'Y: Bitcoin Return', xy=(0.06, 0.22), xycoords='axes fraction',fontsize = 18)


plt.show()
```
