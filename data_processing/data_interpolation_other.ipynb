```python
import pandas as pd
import numpy as np
import os
```

```python
def interpolate_index(file):

  df = pd.read_csv(file)
  df = df.dropna()
  print(df.columns)
  df['date']= pd.to_datetime(df['Data'])
  df = df.set_index("date")
  df  = df['Index'].to_frame()
  df.columns = ['price']
  df.replace(0,-1,inplace=True)
  df2 = df.resample('D').sum()
  df2.replace(0, np.nan, inplace=True)
  df2.replace(-1,0,inplace=True)
  df2 = df2.interpolate(method='linear', limit_direction='forward')
  return df2
```

```python
def add_direction(file):
  df = pd.read_csv(file)
  coinname = file[54:-7]
  df['return'] = df['price'].pct_change()
  df['direction'] = df['return'].apply(lambda x : 1 if x >=0 else -1)
  df.to_csv("/content/drive/MyDrive/S&P Crypto Project/Crypto Data/v3/"+coinname + '_v3.csv')
```
