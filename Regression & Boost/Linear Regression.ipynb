```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import statsmodels.api as sm
from sklearn.metrics import roc_curve, auc
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
df1 = pd.read_csv('1.gt_bitcoin.csv', usecols=[2], skiprows=31, names=['1.gt_bitcoin'])
df2 = pd.read_csv('2.gt_Buy_Bitcoin.csv', usecols=[2], skiprows=31, names=['2.gt_Buy_Bitcoin'])
df3 = pd.read_csv('3.gt_Sell_Bitcoin.csv', usecols=[2], skiprows=31, names=['3.gt_Sell_Bitcoin'])
df4 = pd.read_csv('4.gt_ethereum.csv', usecols=[2], skiprows=31, names=['4.gt_ethereum'])
df5 = pd.read_csv('5.gt_Ukraine_war.csv', usecols=[2], skiprows=31, names=['5.gt_Ukraine_war'])
df6 = pd.read_csv('6.gt_covid.csv', usecols=[2], skiprows=31, names=['6.gt_covid'])
df7 = pd.read_csv('7.s&p_twitter.csv', usecols=[2], skiprows=31, names=['7.s&p_twitter'])
df8 = pd.read_csv('8.CBDC_uncertainty.csv', usecols=[2], skiprows=31, names=['8.CBDC_uncertainty'])
df9 = pd.read_csv('9.CBDC_atten.csv', usecols=[2], skiprows=31, names=['9.CBDC_atten'])
df10 = pd.read_csv('10.news_senti.csv', usecols=[2], skiprows=31, names=['10.news_senti'])
df11 = pd.read_csv('11.fear&greed.csv', usecols=[2], skiprows=31, names=['11.fear&greed'])
df12 = pd.read_csv('12.tweets_pos.csv', usecols=[2], skiprows=31, names=['12.tweets_pos'])
df13 = pd.read_csv('13.tweets_neg.csv', usecols=[2], skiprows=31, names=['13.tweets_neg'])
df14 = pd.read_csv('14.twitter_btccrash_pos.csv', usecols=[2], skiprows=31, names=['14.twitter_btccrash_pos'])
df15 = pd.read_csv('15.twitter_btccrash_neg.csv', usecols=[2], skiprows=31, names=['15.twitter_btccrash_neg'])
```

```python
df_concat = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13, df14, df15], axis=1)
df_remove = df_concat.drop(['2.gt_Buy_Bitcoin', '3.gt_Sell_Bitcoin', '8.CBDC_uncertainty'], axis=1)
# replace "inf" with NaN
df_remove = df_remove.replace('inf', 0)
df_remove = df_remove.replace([np.inf, -np.inf], 0)
df_na = df_remove.fillna(0)
df_na
```

```python
bitcoin = pd.read_csv('bitcoin.csv', usecols=[3], skiprows=31, names=['bitcoin'])
ether = pd.read_csv('ethereum.csv', usecols=[3], skiprows=31, names=['ether'])
BDM_exlarge = pd.read_csv('SP_BDM_exlarge.csv', usecols=[3], skiprows=31, names=['BDM_exlarge'])
BDM = pd.read_csv('SP_BDM.csv', usecols=[3], skiprows=31, names=['BDM'])
```

# 1. Bitcoin

```python
bitcoin_concat = pd.concat([bitcoin,df_na], axis=1)
train, test = train_test_split(bitcoin_concat, test_size = 0.3)
x_train = train.iloc[0:,1:13]
y_train = train.iloc[0:,0]
x_test = test.iloc[0:,1:13]
y_test = test.iloc[0:,0]

# add intercept term to independent variables
x_train = sm.add_constant(x_train)
x_test = sm.add_constant(x_test)

# fit linear regression model on training set
linear_model = sm.OLS(y_train, x_train)
result = linear_model.fit()

# predict outcome variable for the test set
y_pred = result.predict(x_test)

# calculate accuracy
accuracy = r2_score(y_test, y_pred)

# print accuracy
print('Accuracy:', accuracy)
result.summary()
```
